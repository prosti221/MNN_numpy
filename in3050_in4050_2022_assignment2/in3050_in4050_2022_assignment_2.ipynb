### Linear regression
We see that that set (X, t2) is far from linearly separable, and we will explore how various classifiers are able to handle this. We start with linear regression. You may make your own implementation from scratch or start with the solution to the weekly exercise set 7, which we include here.

```python
def add_bias(X):
    # Put bias in position 0
    sh = X.shape
    if len(sh) == 1:
        #X is a vector
        return np.concatenate([np.array([1]), X])
    else:
        # X is a matrix
        m = sh[0]
        bias = np.ones((m,1)) # Makes a m*1 matrix of 1-s
        return np.concatenate([bias, X], axis  = 1) 
```

```python
class NumpyClassifier():
    """Common methods to all numpy classifiers --- if any"""
    
    def accuracy(self,X_test, y_test, **kwargs):
        pred = self.predict(X_test, **kwargs)
        if len(pred.shape) > 1:
            pred = pred[:,0]
        return np.sum(pred==y_test)/len(pred)
    
```

```python
class NumpyLinRegClass(NumpyClassifier):
    def __init__(self):
        self.e = 0                #Number of epochs run
        self.t_loss_hist = []     #Training loss
        self.v_loss_hist = []     #Validation loss
        self.t_accuracy_hist = [] #Training accuracy
        self.v_accuracy_hist = [] #Validation accuracy

    def fit(self, X_train, t_train, eta = 0.01, loss_diff=(1 * 10**-7), validation_set=None):
        """X_train is a Nxm matrix, N data points, m features
        t_train are the targets values for training data"""
        self.reset()
        (k, m) = X_train.shape
        self.weights = weights = np.zeros(m+1)
        XX_train = add_bias(X_train)
        std = -1   #current standard deviation
        self.e = 0 #current epoch
        while std > loss_diff or std == -1:
            #I take the standard deviation of the last 10 loss values
            self.e += 1
            std = np.std(self.t_loss_hist[-10:]) if len(self.t_loss_hist) > 1 else -1
            self.t_loss_hist.append(self.loss(X_train, t_train))
            self.t_accuracy_hist.append(self.accuracy(X_train, t_train))
            if validation_set != None:
                self.v_loss_hist.append(self.loss(*validation_set))
                self.v_accuracy_hist.append(self.accuracy(*validation_set))
            weights -= eta / k *  XX_train.T @ (XX_train @ weights - t_train)
            
    def loss(self, X_test, y_test):
        x = add_bias(X_test)
        return np.average((y_test - x @ self.weights)**2)
    
    def reset(self): #reset values when re-training
        self.e = 0
        self.t_loss_hist = []
        self.v_loss_hist = []
        self.t_accuracy_hist = []
        self.v_accuracy_hist = []
    
    def predict(self, x, threshold=0.5):
        z = add_bias(x)
        score = z @ self.weights
        return score>threshold
```

```python
def logistic(x):
    return 1/(1+np.exp(-x))
```

```python
class NumpyLogReg(NumpyClassifier):
    def __init__(self):
        self.e = 0                #Number of epochs run
        self.t_loss_hist = []     #Training loss
        self.v_loss_hist = []     #Validation loss
        self.t_accuracy_hist = [] #Training accuracy
        self.v_accuracy_hist = [] #Validation accuracy

    def fit(self, X_train, t_train, eta = 0.1, loss_diff=(1 * 10**-4), validation_set=None):
        """X_train is a Nxm matrix, N data points, m features
        t_train are the targets values for training data"""
        self.reset()
        (k, m) = X_train.shape
        XX_train = X_train
        X_train = add_bias(X_train)
        self.weights = weights = np.zeros(m+1)
        std = -1   #current standard deviation
        self.e = 0 #current epoch
        while std > loss_diff or std == -1:
            self.e += 1
            std = np.std(self.t_loss_hist[-10:]) if len(self.t_loss_hist) > 1 else -1
            self.t_loss_hist.append(self.loss(XX_train, t_train))
            self.t_accuracy_hist.append(self.accuracy(XX_train, t_train))
            if validation_set != None:
                self.v_loss_hist.append(self.loss(*validation_set))
                self.v_accuracy_hist.append(self.accuracy(*validation_set))        
            weights -= eta / k *  X_train.T @ (self.forward(X_train) - t_train)      
    
    def forward(self, X):
        return logistic(X @ self.weights)
    
    def score(self, x):
        z = add_bias(x)
        score = self.forward(z)
        return score
    
    def loss(self, X_train, t_train):
        X_train = add_bias(X_train)
        return sum(-np.log(((self.forward(X_train)**t_train)*(1-self.forward(X_train))**(1-t_train))))
    
    def reset(self): #reset values when re-training
        self.e = 0
        self.t_loss_hist = []
        self.v_loss_hist = []
        self.t_accuracy_hist = []
        self.v_accuracy_hist = []
    
    def predict(self, x, threshold=0.5):
        z = add_bias(x)
        score = self.forward(z)
        return (score>threshold).astype('int')
    
    def predict_prob(self, x):
        z = add_bias(x)
        return self.forward(z)
```

### "One-vs-rest" with logistic regression
We saw in the lecture how a logistic regression classifier can be turned into a multi-class classifier using the one-vs-rest approach. We train one logistic regression classifier for each class. To predict the class of an item, we run all the binary classifiers and collect the probability score from each of them. We assign the class which ascribes the highest probability.

Build such a classifier. Train the resulting classifier on (X_train, t_train), test it on (X_val, t_val), tune the hyper-parameters and report the accuracy.

Also plot the decision boundaries for your best classifier similarly to the plots for the binary case.

```python
class RegMultiClass(NumpyClassifier):
    def __init__(self):
        self.sub_classifiers = []
        self.accuracy_val = 0
    
    def fit(self, X_train, t_train, eta = 0.1, loss_diff=(1 * 10**-4), validation_set=None):
        #Transforming training data for 5 seperate classifiers
        tx_train = np.array([[(i==point).astype('int') for point in t_train] for i in range(5)])
        tx_val   = np.array([[(i==point).astype('int') for point in t_val] for i in range(5)])

        #Training the classifiers
        for i in range(5):
            c = NumpyLogReg()
            c.fit(X_train, tx_train[i], eta, loss_diff)
            self.sub_classifiers.append(c)
        
        #Validate on t_val
        if validation_set != None:
            self.accuracy_val = self.accuracy(*validation_set)
            
    def predict(self, x):
        if len(x.shape) > 1:
            res = [np.argmax([sub_cl.predict_prob(val) for sub_cl in self.sub_classifiers]) 
                                                       for val in x]
        else:
            res = np.argmax([sub_cl.predict_prob(x) for sub_cl in self.sub_classifiers])
        return np.array(res)
        
```

First we have to scale our data. Make a standard scaler (normalizer) and scale the data. Remember, not to follow Marsland on this point. The scaler should be constructed from the training data only, but be applied both to training data and later on to validation and test data.

```python
# Your code
def scale(data_set): #Using standard scaling
    data_points, feature_n = data_set.shape;
    #Standard deviation and mean for each feature
    std_i = [np.std(data_set[:,i]) for i in range(feature_n)]
    mean_i = [np.average(data_set[:,i]) for i in range(feature_n)]
    #Normalizing each feature set of X_train
    feature_i_norm = [(data_set[:, i] - mean_i[i]) / std_i[i] for i in range(feature_n)]
    data_set_norm = [[feature_i_norm[i][j] for i in range(feature_n)] 
                                           for j in range(data_points)]
    return np.array(data_set_norm)

X_norm = scale(X_train)
print(np.average(X_norm)) #Should be 0
print(np.std(X_norm))     #Should be 1
```

```python
class MNNClassifier():
    """A multi-layer neural network with one hidden layer"""
    
    def __init__(self,eta = 0.000089, dim_hidden = 128): #eta = 0.0096, hidden=42 best so far
        """Initialize the hyperparameters"""
        self.eta = eta
        self.dim_hidden = dim_hidden
        self.weights1, self.weights2 = (None, None)
        self.accuracies = []    #list of accuracies for the validation data
        self.best_accuracy = -1 #Highest accuracy during training for the validation data
                
    def fit(self, X_train, t_train, epochs = 5000, validation_set=None):
        """Initialize the weights. Train *epochs* many epochs."""
        # Initilaization
        dim_in =  X_train.shape[1]
        dim_out = self.encode(t_train).shape[1]
        self.weights1, self.weights2 = self.init_weights(dim_in, dim_out, self.dim_hidden)
        t_train_enc = self.encode(t_train)
        inp_scaled = self.add_bias((X_train))
        validation_set = (X_train, t_train) if validation_set is None else validation_set
        
        best_weights = [-1, None] #Array with accuracy and copy of the best weights so far for the validation data
        for e in range(epochs):
            #Run the forward pass
            a_out, y_out = self.forward(inp_scaled)
            #Check accuracy
            if ((acc := self.accuracy(*validation_set, y_out=None)) > best_weights[0]):
                best_weights = [acc, (np.copy(self.weights1), np.copy(self.weights2))]
                self.best_accuracy = acc
            self.accuracies.append(acc)
            print(acc)
            #delta terms
            delta_0 = (y_out - t_train_enc) * y_out * (1 - y_out)
            delta_1 = a_out*(1 - a_out) * np.dot(delta_0, self.weights2.T)
            #Update weights
            self.weights2 -= self.eta * np.dot(a_out.T, delta_0) 
            self.weights1 -= self.eta * np.dot(inp_scaled.T, delta_1[: , :-1])
        #Assign the weights to the ones with the best accuracy
        self.weights1, self.weights2 = best_weights[1]
            
    def forward(self, X):
        """Perform one forward step. 
        Return a pair consisting of the outputs of the hidden_layer
        and the outputs on the final layer"""
        a_in = X @ self.weights1
        a_out = self.sigmoid(a_in)
        y_in = a_out_bias = self.add_bias(a_out)
        y_in = y_in @ self.weights2
        y_out = self.sigmoid(y_in)
        return (a_out_bias, y_out)
        
    def predict(self, X_train):
        return np.argmax(self.forward(self.add_bias((X_train)))[1], axis=1)
    
    def score(self, X_train):
        return self.forward(self.add_bias(self.scale(X_train)))[1][:, 1]
    
    def accuracy(self, X_train, t_test, y_out=None):
        """Calculate the accuracy of the classifier for the pair (X_test, t_test)
        Return the accuracy"""
        if y_out is not None:
            acc = [int(np.argmax(y_out[i]) == t_test[i]) for i in range(t_test.shape[0])]
        else:
            inp = self.add_bias((X_train))
            acc = [int(np.argmax(self.forward(inp[i])[1]) == t_test[i]) for i in range(t_test.shape[0])]
        return sum(acc) / len(acc)
        #Fill in the code
        
    def scale(self, data_set): #Using standard scaling
        data_points, feature_n = data_set.shape;
        #Standard deviation and mean for each feature
        std_i = [np.std(data_set[:,i]) for i in range(feature_n)]
        mean_i = [np.average(data_set[:,i]) for i in range(feature_n)]
        #Normalizing each feature set of X_train
        feature_i_norm = [(data_set[:, i] - mean_i[i]) / std_i[i] for i in range(feature_n)]
        data_set_norm = [[feature_i_norm[i][j] for i in range(feature_n)] 
                                               for j in range(data_points)]
        return np.array(data_set_norm)
    
    def encode(self, data_set): #Converts output to one-hot encoding
        class_n = np.max(data_set) + 1; data_points = len(data_set)
        return np.array([[np.array((data_set[i] == j)).astype('int') for j in range(class_n)] 
                                                                     for i in range(data_points)]) 
    def init_weights(self, dim_in, dim_out, dim_hidden):
        w1 = np.random.randn(dim_in + 1, dim_hidden)/np.sqrt(dim_in+1 * dim_hidden)
        w2 = np.random.randn(dim_hidden + 1, dim_out)/np.sqrt(dim_hidden+1 * dim_out)
        return (w1, w2)
    
    def sigmoid(self, X):
        return 1.0/(1.0+np.exp(-X))
    
    def add_bias(self, X):
        # Put bias in position 0
        sh = X.shape
        if len(sh) == 1:
            #X is a vector
            return np.concatenate([np.array([-1]), X])
        else:
            # X is a matrix
            m = sh[0]
            bias = -np.ones((m,1)) # Makes a m*1 matrix of 1-s
            return np.concatenate([bias, X], axis  = 1) 
```
